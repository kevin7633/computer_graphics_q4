```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


img = cv2.imread('/content/캡sdfe처.PNG', cv2.IMREAD_GRAYSCALE)


if img is None:
    print("Error: Image could not be loaded. Please check the file path.")
else:

    _, binary = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)


    kernel = np.ones((5,5), np.uint8)


    dilation = cv2.dilate(binary, kernel, iterations=1)
    erosion = cv2.erode(binary, kernel, iterations=1)

    opening = cv2.dilate(cv2.erode(binary, kernel, iterations=1), kernel, iterations=1)

    closing = cv2.erode(cv2.dilate(binary, kernel, iterations=1), kernel, iterations=1)


    plt.figure(figsize=(15,4))

    plt.subplot(1,5,1)
    plt.title("Original")
    plt.imshow(binary, cmap='gray')
    plt.axis('off')

    plt.subplot(1,5,2)
    plt.title("Dilation")
    plt.imshow(dilation, cmap='gray')
    plt.axis('off')

    plt.subplot(1,5,3)
    plt.title("Erosion")
    plt.imshow(erosion, cmap='gray')
    plt.axis('off')

    plt.subplot(1,5,4)
    plt.title("Opening")
    plt.imshow(opening, cmap='gray')
    plt.axis('off')

    plt.subplot(1,5,5)
    plt.title("Closing")
    plt.imshow(closing, cmap='gray')
    plt.axis('off')

    plt.show()


    orig_white = np.sum(binary == 255)
    dil_white = np.sum(dilation == 255)
    ero_white = np.sum(erosion == 255)
    op_white = np.sum(opening == 255)
    cl_white = np.sum(closing == 255)

    print(f"Original white pixels: {orig_white}")
    print(f"Dilation white pixels: {dil_white}")
    print(f"Erosion white pixels: {ero_white}")
    print(f"Opening white pixels: {op_white}")
    print(f"Closing white pixels: {cl_white}")
```
